```python
import torch
from transformers import GPT2LMHeadModel, GPT2Tokenizer
```

```python
# Load pre-trained GPT-2 model and tokenizer
model = GPT2LMHeadModel.from_pretrained("gpt2")
tokenizer = GPT2Tokenizer.from_pretrained("gpt2")
```

```python
# Set the model to evaluation mode
model.eval()
```

```python
# Define a function to generate code snippets given a prompt
def generate_code(prompt, max_length=50, temperature=0.7):
    input_ids = tokenizer.encode(prompt, return_tensors="pt")
    output = model.generate(input_ids, max_length=max_length, temperature=temperature, pad_token_id=tokenizer.eos_token_id)
    generated_code = tokenizer.decode(output[0], skip_special_tokens=True)
    return generated_code
```

```python
# Example prompt
prompt = "Write a python method using def"
# prompt = "this calculates the mean"

# Generate code snippet given the prompt
generated_code = generate_code(prompt)

# Print the generated code snippet
print("Generated Code:")
print(generated_code)
```
